## Annual growth of imports of goods and services
Data from Our World in Data website

```python
# Import libraries
import pandas as pd
# libraries for visualization
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('setup complete')
```

Read file

```python
#import the csv data
imports = pd.read_csv(r"C:\Users\CYPHA\OneDrive\Documents\VisualizeGhana\annual-growth-of-imports-of-goods-and-services.csv", sep=',')
imports.head()
```

Get data related to Ghana only

```python
imports_ghana = imports.loc[imports['Entity']== 'Ghana']
imports_ghana
```

Reset the index to be the year

```python
# Reset the index to year
imports_ghana = imports_ghana.set_index('Year')
imports_ghana
```

Rename columns

```python
# # Rename column
imports_ghana = imports_ghana.rename(columns = {'17.13.1 - Annual growth of imports of goods and services (%) - NE_IMP_GNFS_KD_ZG': 'Annual Imports'})
imports_ghana
```

Drop unwanted columns

```python
imports_ghana = imports_ghana.drop(columns={'Entity', 'Code'})
imports_ghana
```

Save results to CSV file

```python
# Save to a new csv file
imports_CSV = imports_ghana.to_csv("annual-growth-of-imports-of-goods-and-services-ghana.csv", index=False)
imports_CSV
```

## Visualization

First get the range needed for the visualization (in my case last seven years)

```python
# Data for Imports visualization
imports_ghana_viz = imports_ghana.loc[2015:2021]
imports_ghana_viz
```

```python
# Data for imports amount visualization
imports_amt_viz = imports_amt.loc[2015:2021]
imports_amt_viz
```

```python
# Data for exports visualization
exports_viz = exports_ghana.loc[2015:2021]
exports_viz
```

Plot the values

```python
# Creat area for plots
fig, axs = plt.subplots(3,1, figsize=(10,17))
# Plot for respective data
# ax=axs[0] specifies the location to be ploted
sns.lineplot(x=exports_viz.index, y=exports_viz['Annual Exports'], marker='o', ax=axs[0]) 
axs[0].set_title('Annual Growth of Exports')

sns.lineplot(x=imports_ghana_viz.index, y=imports_ghana_viz['Annual Imports'], marker='o', ax=axs[1])
axs[1].set_title("Annual Growth of imports")

sns.lineplot(x=imports_amt_viz.index, y=imports_amt_viz['Imports'], marker='o', ax=axs[2])
axs[2].set_title('Amount Spent on Imports')
```
